```python
def linear_regression(x, y):
    n = len(x)

    # calculate means
    x_bar = sum(x) / n
    y_bar = sum(y) / n

    # calculate numerator and denominator
    num = 0  # Sxy
    den = 0  # Sxx
    for i in range(n):
        num += (x[i] - x_bar) * (y[i] - y_bar)
        den += (x[i] - x_bar) ** 2

    m = num / den
    c = y_bar - m * x_bar
    return c, m
```

```python
def ridge_regression(x, y, lam):
    n = len(x)
    x_bar = sum(x) / n
    y_bar = sum(y) / n

    num = 0
    den = 0
    for i in range(n):
        num += (x[i] - x_bar) * (y[i] - y_bar)
        den += (x[i] - x_bar) ** 2

    m = num / (den + lam)
    c = y_bar - m * x_bar
    return c, m
```

```python
def lasso_regression(x, y, lam):
    n = len(x)
    x_bar = sum(x) / n
    y_bar = sum(y) / n

    num = 0
    den = 0
    for i in range(n):
        num += (x[i] - x_bar) * (y[i] - y_bar)
        den += (x[i] - x_bar) ** 2

    # soft thresholding
    if abs(num) <= lam:
        m = 0
    else:
        m = (abs(num) - lam) / den
        if num < 0:
            m = -m

    c = y_bar - m * x_bar
    return c, m
```

```python
def predict(c, m, x):
    return [c + m * xi for xi in x]
```

```python
def mse(actual, predicted):
    n = len(actual)
    s = 0
    for i in range(n):
        s += (actual[i] - predicted[i]) ** 2
    return s / n
```

```python
x = [2, 4, 6, 8, 10]
y = [5, 8, 9, 13, 16]

# Linear Regression
c_lin, m_lin = linear_regression(x, y)
p_lin = predict(c_lin, m_lin, x)
print("Linear Regression:")
print("c =", c_lin, "m =", m_lin)
print("Predicted =", p_lin)
print("MSE =", mse(y, p_lin))
```

```python
# Ridge Regression
c_rid, m_rid = ridge_regression(x, y, 1)
p_rid = predict(c_rid, m_rid, x)
print("\nRidge Regression:")
print("c =", c_rid, "m =", m_rid)
print("Predicted =", p_rid)
print("MSE =", mse(y, p_rid))
```

```python
# Lasso Regression
c_las, m_las = lasso_regression(x, y, 1)
p_las = predict(c_las, m_las, x)
print("\nLasso Regression:")
print("c =", c_las, "m =", m_las)
print("Predicted =", p_las)
print("MSE =", mse(y, p_las))
```
